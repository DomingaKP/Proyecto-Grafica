 Suma de los promedios mensuales de toallas desechables más ácido mefenámino para un persona con un ciclo de 5 días para los años 2021, 2022 y 2023:

```python
# Instalar las librerías necesarias
!pip install altair_viewer

# Importar librerías
import pandas as pd
import altair as alt

# Crear los datos según la tabla
data = {
    "Producto": ["Toallas Desechables", "Toallas Desechables", "Toallas Desechables",
                 "Ácido Mefenámico", "Ácido Mefenámico", "Ácido Mefenámico"],
    "Año": [2021, 2022, 2023, 2021, 2022, 2023],
    "Promedio Mensual (5 días)": [3.167, 5.040, 5.532, 2.720, 2.427, 2.350]
}

# Crear el DataFrame
df = pd.DataFrame(data)

# Calcular la suma de los promedios mensuales para cada año
suma_promedios = df.groupby("Año")["Promedio Mensual (5 días)"].sum().reset_index()
suma_promedios.rename(columns={"Promedio Mensual (5 días)": "Suma Promedios Mensuales"}, inplace=True)

# Crear el gráfico de barras
barras = alt.Chart(suma_promedios).mark_bar().encode(
    x=alt.X("Año:O", title="Año"),
    y=alt.Y("Suma Promedios Mensuales:Q", title="Suma de Promedios Mensuales"),
    color=alt.Color("Año:O", scale=alt.Scale(scheme="plasma")),
    tooltip=[
        alt.Tooltip("Año:O", title="Año"),
        alt.Tooltip("Suma Promedios Mensuales:Q", title="Suma Promedios", format="$.3f")
    ]
)

# Agregar etiquetas de texto con los valores
etiquetas = barras.mark_text(
    align="center",
    baseline="middle",
    dy=-10  # Ajusta la posición del texto sobre la barra
).encode(
    text=alt.Text("Suma Promedios Mensuales:Q", format="$.3f")  # Formato con 3 decimales y signo de peso
)

# Combinar las barras y las etiquetas
grafico = (barras + etiquetas).properties(
    title="Suma de Promedios Mensuales (Toallas Desechables + Ácido Mefenámico, Ciclo 5 Días)",
    width=600,
    height=400
)

# Mostrar el gráfico
grafico
```
