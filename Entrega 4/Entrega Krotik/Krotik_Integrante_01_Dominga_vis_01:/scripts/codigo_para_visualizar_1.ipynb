```python
# Instalar Altair y Vega Datasets
!pip install altair vega_datasets

import pandas as pd
import altair as alt
from vega_datasets import data

# Cargar el archivo CSV y limpiar los datos como en los pasos anteriores
file_path = '/content/DATOS LIMPIOS REGIONES 2024.csv'
data = pd.read_csv(file_path)
data.columns = data.iloc[0]
data = data.drop(0)
data = data[['REGION', 'SEXO', 'ATENDIDOS']]
data['ATENDIDOS'] = pd.to_numeric(data['ATENDIDOS'], errors='coerce')
data_female = data[data['SEXO'] == 'Mujer']
data_female_region = data_female.groupby('REGION')['ATENDIDOS'].sum().reset_index()

# Crear gráfico de barras con sumatoria de mujeres en cárceles por región, ordenado
chart = alt.Chart(data_female_region).mark_bar().encode(
    x='sum(ATENDIDOS):Q',  # Sumar la cantidad de mujeres en cada región
    y=alt.Y('REGION:N', sort='-x')  # Ordenar regiones de mayor a menor
).properties(
    title='Regiones con Más Mujeres en Cárceles'
)

chart
```

```python
# Instalar Altair y Vega Datasets
!pip install altair vega_datasets

import pandas as pd
import altair as alt
from vega_datasets import data

# Cargar el archivo CSV y limpiar los datos
file_path = '/content/DATOS LIMPIOS REGIONES 2024.csv'
data = pd.read_csv(file_path)
data.columns = data.iloc[0]
data = data.drop(0)
data = data[['REGION', 'SEXO', 'ATENDIDOS']]
data['ATENDIDOS'] = pd.to_numeric(data['ATENDIDOS'], errors='coerce')

# Filtrar para solo hombres
data_male = data[data['SEXO'] == 'Hombre']

# Agrupar por región y sumar los valores de 'ATENDIDOS' para cada región
data_male_region = data_male.groupby('REGION')['ATENDIDOS'].sum().reset_index()

# Crear gráfico de barras con el título personalizado para el eje X
chart = alt.Chart(data_male_region).mark_bar().encode(
    x=alt.X('sum(ATENDIDOS):Q', title='Número de Hombres Atendidos'),  # Cambia el título del eje X
    y=alt.Y('REGION:N', sort='-x')  # Ordena las regiones de mayor a menor
).properties(
    title='Regiones con Más Hombres en Cárceles'
)

chart
```

```python
# Importar librerías
import pandas as pd
import altair as alt

# Cargar el archivo CSV
file_path = '/content/Copia de DATOS LIMPIOS REGIONES 2024.csv'
data = pd.read_csv(file_path)

# Renombrar las columnas para facilitar su uso
data.columns = ['Fecha', 'Region', 'Establecimiento', 'Administracion', 'Subsistema',
                'RegimenControl', 'TipoPoblacion', 'Poblacion', 'Sexo', 'Atendidos']

# Asegurarse de que la columna 'Atendidos' sea numérica
data['Atendidos'] = pd.to_numeric(data['Atendidos'], errors='coerce')

# Filtrar datos necesarios
data_filtered = data[['Region', 'Subsistema', 'Sexo', 'Atendidos']].dropna()

# Visualización 1: Número total de atendidos por subsistema
chart1 = alt.Chart(data_filtered).mark_bar().encode(
    x=alt.X('sum(Atendidos):Q', title='Total Atendidos'),
    y=alt.Y('Subsistema:N', sort='-x', title='Subsistema'),
    color='Subsistema:N'
).properties(
    title='Total de Atendidos por Subsistema',
    width=600,
    height=400
)

# Visualización 2: Número de atendidos por región y subsistema
chart2 = alt.Chart(data_filtered).mark_bar().encode(
    x=alt.X('sum(Atendidos):Q', title='Total Atendidos'),
    y=alt.Y('Region:N', sort='-x', title='Región'),
    color='Subsistema:N',
    tooltip=['Subsistema', 'sum(Atendidos)']
).properties(
    title='Atendidos por Región y Subsistema',
    width=700,
    height=500
)

# Visualización 3: Distribución de sexo por subsistema
chart3 = alt.Chart(data_filtered).mark_bar().encode(
    x=alt.X('sum(Atendidos):Q', title='Total Atendidos'),
    y=alt.Y('Subsistema:N', sort='-x', title='Subsistema'),
    color='Sexo:N',
    tooltip=['Sexo', 'sum(Atendidos)']
).properties(
    title='Distribución de Sexo por Subsistema',
    width=600,
    height=400
)

# Mostrar las visualizaciones
chart1 & chart2 & chart3
```
